```python
#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv file
df = pd.read_csv("youtube.csv", encoding= 'unicode_escape')
```

```python
#Show records in dataframe
df.head()
```

```python
#Create a subset of dataset to show top 1000 records
df_onethousand_rows=df[0:1000]
df_onethousand_rows.head()
```

```python
#Show the number of records of new subset dataframe
df_onethousand_rows.shape
```

```python
#Distribution of channeltype 
df_onethousand_rows.channeltype.value_counts()
```

```python
#Visualize distribution of channeltype by videoviews
df_onethousand_rows.groupby(['channeltype'])['videoviews'].count().plot(kind='bar')
```
